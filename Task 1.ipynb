```python
# Cohort analysis
import datetime as dt
#function for the month
def get_month(x):
    return dt.datetime(x.year, x.month,1)
#apply the function
df_transactions['InvoiceMonth'] = df_transactions['DATE'].apply(get_month)
df_transactions.tail()
```

```python
#create a column index withthe date aka 1st time customer was aquired
df_transactions['Cohort_Month'] = df_transactions.groupby('LYLTY_CARD_NBR')['InvoiceMonth'].transform(min)
df_transactions.head(20)
```

```python
#create  a date element function to get a series for subtraction
def get_elements(df,column):
    day = df[column].dt.day
    month = df[column].dt.month
    year = df[column].dt.year
    return day,month,year
    
```

```python
# get date slements for our cohort and invoice columns
_,Invoice_month,Invoice_year= get_elements(df_transactions,'InvoiceMonth')
_,Cohort_month,Cohort_year= get_elements(df_transactions,'Cohort_Month')
```

```python
#check the series
Cohort_year[:5]
```

```python
#create a cohort  index
year_diff = Invoice_year - Cohort_year
month_diff = Invoice_month - Cohort_month
df_transactions['Cohort_index'] = year_diff*12 + month_diff +1
df_transactions.head(10)
```

```python
#Count the customer id by grouping by cohort month and index
cohort_data = df_transactions.groupby(['Cohort_Month','Cohort_index'])['LYLTY_CARD_NBR'].apply(pd.Series.nunique).reset_index()
#df_transactions.groupby(['Cohort_Month','Cohort_index'])['LYLTY_CARD_NBR'].nunique()
```

```python
# create pivot table
cohort_table = cohort_data.pivot(index ='Cohort_Month',columns=['Cohort_index'],values = 'LYLTY_CARD_NBR')
cohort_table
```

```python
# Change index
#cohort_table.index = cohort_table.index.strftime('%B,%Y')
#visualize our results in a heatmap
plt.figure(figsize =(21,10))
sns.heatmap(cohort_table,annot = True,cmap ='Blues' )
```

```python
#Cohort table for percentage
new_cohort_table = cohort_table.divide(cohort_table.iloc[:,0],axis = 0)
new_cohort_table
```

```python
#create a percentage
plt.figure(figsize =(21,10))
sns.heatmap(new_cohort_table,annot = True,cmap ='coolwarm',fmt='.0%' )
```
